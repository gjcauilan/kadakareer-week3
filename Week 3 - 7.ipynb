```python
import boto3
import pandas as pd
from trp import Document

bucketname = "kadakareer1"
tablename = "Annex-G_Psycho-Evaluation-Form 1.jpg"

module = boto3.client('textract')

response = module.analyze_document(
    Document={
        'S3Object': {
            'Bucket': bucketname,
            'Name': tablename
        }
    },
    FeatureTypes=["TABLES"]
)
doc = Document(response)

table_data = []

for page in doc.pages:
    for table in page.tables:
        for row in table.rows:
            row_data = []
            for cell in row.cells:
                row_data.append(cell.text)
            table_data.append(row_data)

df = pd.DataFrame(table_data)

print('-----Amazon Textract has detected as follows-----')
print(df)

df.to_csv('Week 3 - 7.csv', index=False)
```
